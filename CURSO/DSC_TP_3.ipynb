Ejemplo 1: Desarrollar y calcular la confiabilidad de un sistema de dos componentes en
paralelos para una vida útil de y que se supone que son independientes y
distribuidas exponencialmente con tasas de falla , .

```python
## Importación de libreria
import numpy as np
import matplotlib.pyplot as plt
```

```python
## Asignación de valores
lambda1 = np.array([[0.0002], [0.0004], [0.0001]]) # Failure rate 1
lambda2 = np.array([[0.0008], [0.0006], [0.0001]])# Failure rate 2
x_year = 24*365 # cantidad de horas al año
x_years = 30*x_year # cantidad de horas al año
N = 1000
x = np.linspace(0, x_years, N)
# Calculo de confiabilidad e inconfiabilidad
Rs = np.exp(-lambda1*x)+np.exp(-lambda2*x)-np.exp(-(lambda1+lambda2)*x) # Confiabil
Fs = 1 - Rs
index_2years = int(x_year/(x_years/N))*2
plt.plot(x[:index_2years], Rs[0][:index_2years], label="$Rs_1$")
plt.plot(x[:index_2years], Rs[1][:index_2years], label="$Rs_2$")
plt.plot(x[:index_2years], Rs[2][:index_2years], label="$Rs_3$")
plt.axvline(x=x_year, color = 'm')
plt.legend()
plt.grid()
plt.show()
```

```python
# Calculo MTTF
MTTF_years = np.trapz(Rs, x=x)
print(f"MTTF calculado en python = {MTTF_years}")
MTTF_teorico = 1/lambda1 + 1/lambda2 - 1/(lambda1+lambda2)
print(f"MTTF calculado teoricamente = {MTTF_teorico.T}")
# Calculo la tasa de falla del sistema
numerador_Zs = lambda1*np.exp(-lambda1*x)+lambda2*np.exp(-lambda2*x)-\
(lambda1+lambda2)*np.exp(-(lambda1+lambda2)*x)
Zs = numerador_Zs / Rs
plt.plot(x[:index_2years], Zs[0][:index_2years], label="$Zs_1$ $\mu_1=0.0002\mu_2=0.0008$")
plt.plot(x[:index_2years], Zs[1][:index_2years], label="$Zs_2$ $\mu_1=0.0004\mu_2=0.0006$")
plt.plot(x[:index_2years], Zs[2][:index_2years], label="$Zs_3$ $\mu_1=0.0001\mu_2=0.0001$")
plt.axvline(x=x_year, color = 'm')
plt.legend()
plt.grid()
plt.show()
```

En la figura anterior se mostraba la tasa de fallo del sistema  para dos componentes independeintes.

```python
# Calculo MTTF
MTTF_years = np.trapz(Rs, x=x)
print(f"MTTF calculado en python = {MTTF_years}")
MTTF_teorico =  1/(lambda1+lambda2) + 1/(lambda1+lambda3) + 1/(lambda2+lambda3) - 2/(lambda1+lambda2+lambda3)
print(f"MTTF calculado teoricamente = {MTTF_teorico.T}")
# Calculo la tasa de falla del sistema
numerador_Zs = (lambda1+lambda2)*np.exp(-(lambda1+lambda2)*x)+(lambda1+lambda3)*np.exp(-(lambda1+lambda3)*x)+(lambda2+lambda3)*np.exp(-(lambda2+lambda3)*x)-2*(lambda1+lambda2+lambda3)*np.exp(-(lambda1+lambda2+lambda3)*x)

Zs = numerador_Zs / Rs
plt.plot(x[:index_2years], Zs[0][:index_2years], label="$Zs_1$ $\mu_1=0.00000001$  $\mu_2=0.00000001$  $\mu_3=0.000000002$")
plt.plot(x[:index_2years], Zs[1][:index_2years], label="$Zs_2$ $\mu_1=0.000000012$ $\mu_2=0.000000012$ $\mu_3=0.000000002$")
plt.plot(x[:index_2years], Zs[2][:index_2years], label="$Zs_3$ $\mu_1=0.000000015$ $\mu_2=0.000000015$ $\mu_3=0.000000002$")
plt.axvline(x=x_year, color = 'm')
plt.legend()
plt.grid()
plt.show()
```
